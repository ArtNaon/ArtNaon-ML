```python
class_names = ['Abstract', 'Academic_Art', 'Art_Nouveau', 'Baroque', 'Expressionism', 'Japanese_Art', 'Neoclassicism',
                'Primitivism', 'Realism', 'Renaissance', 'Rococo', 'Romanticism', 'Symbolism']
```

```python
uploaded = files.upload()

for fn in uploaded.keys():

  #Upload image
  path = fn
  img = load_img(path, target_size=(224, 224))
  x = img_to_array(img)
  x = np.expand_dims(x, axis=0)

  #Predict image
  images = np.vstack([x])
  classes = model.predict(images, batch_size=32)
  print(fn)
  print(classes)
```

```python
predicted_classes = np.argmax(classes, axis=-1)
predicted_labels = [class_names[class_index] for class_index in predicted_classes]
# Print predicted labels
for label in predicted_labels:
    print(label)
```

```python
import os
import matplotlib.pyplot as plt

base_path = '/content/drive/MyDrive/Dataset/DatasetV2/WikiArt Art Movements Styles/Arts'

file_path = base_path + '/' + label
```

```python
num_images_to_display = 9
counter = 0

fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes = axes.flatten()

# Traverse the directory structure
for root, dirs, files in os.walk(file_path):
    for file in files:
        # Check if the file is an image
        if file.endswith('.jpg') or file.endswith('.png'):
            # Construct the absolute file path
            file_path = os.path.join(root, file)

            # Load and display the image
            image = plt.imread(file_path)

            # Show the image on the next available subplot
            axes[counter].imshow(image)
            axes[counter].axis('off')
            counter += 1

            # Check if the desired number of images has been displayed
            if counter == num_images_to_display:
                break

    if counter == num_images_to_display:
        break

# Hide remaining empty subplots
for i in range(counter, len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
